```python
#practical No.2
import nltk
nltk.download("book")
from nltk.corpus import wordnet
synonyms=[]
antonyms=[]
for syn in wordnet.synsets(input("enter a word")):
  for l in syn.lemmas():
    synonyms.append(l.name())
    if l.antonyms():
      antonyms.append(l.antonyms()[0].name())
print(set(synonyms))
print(set(antonyms))
```
